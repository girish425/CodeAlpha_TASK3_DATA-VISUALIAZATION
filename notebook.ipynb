```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel("task3.xlsx")
```

```python
df.head()
df.shape
df.info()
```

```python
df_clean = df.loc[:, ~df.columns.str.contains("Unnamed")]
```

```python
plan_counts = df_clean["Subscription_Plan"].value_counts()
```

```python
country_counts = df_clean["Country"].value_counts().head(5)
```

```python
avg_satisfaction = df_clean.groupby("Subscription_Plan")["Satisfaction_Score"].mean()
```

```python
#BAR GRAPH
plt.figure()
plan_counts.plot(kind="bar")
plt.title("Subscription Plan Distribution")
plt.xlabel("Subscription Plan")
plt.ylabel("Number of Users")
plt.show()
```

```python
#BAR GRAPH
plt.figure()
country_counts.plot(kind="bar")
plt.title("Top 5 Countries by Number of Users")
plt.xlabel("Country")
plt.ylabel("Users")
plt.show()
```

```python
#BAR GRAPH
plt.figure()
avg_satisfaction.plot(kind="bar")
plt.title("Average Satisfaction Score by Subscription Plan")
plt.xlabel("Subscription Plan")
plt.ylabel("Average Satisfaction Score")
plt.show()
```

```python
#SCATTER PLOT
plt.figure(figsize=(8,6))

plt.scatter(df["Satisfaction_Score"], df["Annual_Income_INR"])

plt.title("Satisfaction Score vs Annual Income")
plt.xlabel("Satisfaction Score")
plt.ylabel("Annual Income (INR)")

plt.show()
```

```python
#HISTOGRAM 
plt.figure(figsize=(8,6))
plt.hist(df['Age'], bins=10, edgecolor='black')
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.title("Histogram of Age")
plt.show()
```

```python
#PIE CHART
plan_counts = df["Subscription_Plan"].value_counts()

plt.figure()
plt.pie(plan_counts, labels=plan_counts.index, autopct='%1.1f%%')
plt.title("Subscription Plan Share")
plt.show()
```
